```python
## grouped bars

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations, [0 1 2 3 4]
width = 0.25  # the width of the bars

fig, ax = plt.subplots()
#fig = plt.figure()
#ax = plt.axes()

rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# # Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend();

# plt.show()
```

```python
## impostare stile sulla griglia, e etichette sulle barre

fig, ax = plt.subplots()

ax.grid(linestyle='--', color='0.75', axis = 'y')  # impostare proprietà della griglia
ax.set_axisbelow(True)    # Grid behind bars

rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# # Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)

ax.legend()

# Add bar labels
ax.bar_label(rects1, padding=3) # bar_label exists in matplotlib versions > 3.4
ax.bar_label(rects2, padding=3) 

ax.set_ylim(0,40);
```

```python
## stacked bars

fig, ax = plt.subplots()
ax.grid(linestyle='--', color='0.75', axis = 'y');

ax.set_axisbelow(True)    # Grid behind bars

p1 = ax.bar(labels, men_means, width, label='Men')
p2 = ax.bar(labels, women_means, width, bottom=men_means,
       label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

# Label with label_type 'center' instead of the default 'edge'
ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2)
ax.set_ylim(0,70);
```

```python
## uso misto di subplot, marker, etc.
## ref. https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers

fig, axs = plt.subplots(2, 3, sharex=True, sharey=True, figsize=(16,12));

# marker symbol
axs[0, 0].scatter(x, y, s=80, marker=">")
axs[0, 0].set_title("marker='>'")
# marker from TeX
axs[0, 1].scatter(x, y, s=80, marker=r'$\alpha$')
axs[0, 1].set_title(r"marker=r'\$\alpha\$'")

# marker from path
verts = [[-1, -1], [1, -1], [1, 1], [-1, -1]]
axs[0, 2].scatter(x, y, s=80, marker=verts)
axs[0, 2].set_title("marker=verts")

# regular polygon marker
axs[1, 0].scatter(x, y, s=80, marker=(5, 0))
axs[1, 0].set_title("marker=(5, 0)")

# regular star marker
axs[1, 1].scatter(x, y, s=80, marker=(5, 1))
axs[1, 1].set_title("marker=(5, 1)")

# regular asterisk marker
axs[1, 2].scatter(x, y, s=80, marker=(5, 2))
axs[1, 2].set_title("marker=(5, 2)");
```

```python
# Notched boxplot

fig, ax = plt.subplots()
ax.boxplot(data, 1)
ax.set_title('Notched Box Plot');
```

```python
# Notched boxplot without outliers

fig, ax = plt.subplots()
ax.boxplot(data, 1, '')
ax.set_title('Notched Box Plot without Outliers');
```

```python
# Notched boxplot without outliers

fig, ax = plt.subplots()
ax.boxplot(data, 0, 'gD')
ax.set_title('Boxplot with different symbol for outliers');
```

```python
## ESEMPIO 1. IRIS

import sklearn 
import pandas as pd 
import numpy as np

from sklearn import datasets

# Load the iris dataset
iris = datasets.load_iris()

# Load it to a pandas dataframe:
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['label'])

iris_df['label_names'] = iris_df['label'].apply(lambda x: iris.target_names[int(x)])

# Let's have a look at how the data looks like
display(iris_df)
```

```python
# Visualize the data using Matplotlib
import matplotlib.pyplot as plt

colours = ['orange', 'blue', 'green']
species = iris.target_names

f = plt.figure(figsize=(8,6))

for i in range(0, 3):    
    species_df = iris_df[iris_df['label'] == i]    
    plt.scatter(        
        species_df['petal length (cm)'],        
        species_df['sepal length (cm)'],
        color=colours[i],        
        alpha=0.5,        
        label=species[i]   
    )
plt.xlabel('petal length (cm)')
plt.ylabel('sepal length (cm)')
plt.title('Iris dataset: petal length vs sepal length')
plt.legend(loc='lower right')
plt.show()
```

```python
# Visualize the data using Seaborn
# Matrix scatter plot. 

import seaborn as sns

sns.set_style("darkgrid")
sns.pairplot(iris_df[iris.feature_names + ['label_names']], hue="label_names",height=3);
```

```python
#### TIME SERIES ANALYSIS example

"""
    Dataset: Temperature massime medie di Gennaio a New York City nel periodo 1895–2022. 
    Link: https://www.ncdc.noaa.gov/cag/city/time-series/USH00300042-tavg-1-5-1895-2022.csv?base_prd=true&begbaseyear=1901&endbaseyear=2000
    Obiettivo: Utilizzare il metodo della regressione lineare per predire l’andamento delle temperature nel futuro    
"""
```

```python
import pandas as pd
#nyc = pd.read_csv('USH00305801-tmax-1-1-1895-2022.csv') # se non uso skiprows prendo delle righe di header che non mi interessano

nyc = pd.read_csv('USH00305801-tmax-1-1-1895-2022.csv', skiprows = 4) # skiprows mi permette di bypassare righe che non mi interessano

## alternativamente mettere direttamente il link
# nyc = pd.read_csv('https://www.ncdc.noaa.gov/cag/city/time-series/USH00300042-tavg-1-5-1895-2022.csv?base_prd=true&begbaseyear=1901&endbaseyear=2000',
#                     skiprows=4)
nyc
```

```python
# Cleaning the Data
nyc.columns = ['Date', 'Temperature', 'Anomaly']
nyc.head()
```

```python
nyc.tail()
```

```python
nyc.Date = nyc.Date.floordiv(100) # remove 01 from Date

nyc.tail()
```

```python
# Calculating Basic Descriptive Statistics for the Dataset
pd.set_option('precision', 2)
nyc.describe()
```

```python
# Forecasting Future January Average High Temperatures

from scipy import stats

linear_regression = stats.linregress(x=nyc.Date, y=nyc.Temperature)

linear_regression.slope
```

```python
linear_regression.intercept
```

```python
# y = mx + b, to predict the average January temperature in New York City for a given year. 
# Let’s predict the average Fahrenheit temperature for January of 2019. => x = 2019

linear_regression.slope * 2019 + linear_regression.intercept # nel 2019

```

```python
linear_regression.slope * 1890 + linear_regression.intercept # nel 1890
```

```python
linear_regression.slope * 2023 + linear_regression.intercept # predizione nel 2023
```

```python
## Calcoliamo la retta di regressione e plottiamo i dati insieme utilizzando scatter e plot

from matplotlib import pyplot as plt

fig, ax = plt.subplots()

y = linear_regression.slope * nyc.Date + linear_regression.intercept # regression line


ax.scatter(nyc.Date, nyc.Temperature); # plot temp vs date
ax.plot(nyc.Date, y) # plot regressione line


type(y)
```

```python
"""
Utilizziamo la funzione regplot di Seaborn per tracciare ciascun punto dati 
con le date sull'asse x e le temperature sull'asse y. 
La funzione regplot crea il grafico a dispersione o scattergram 
sottostante in cui i punti blu sparsi rappresentano le temperature 
per le date date e la linea retta visualizzata attraverso i punti è la linea di regressione

For readers with more statistics background, the shaded area surrounding the regression 
line is the 95% confidence interval for the regression line
"""

import seaborn as sns
sns.set_style('whitegrid')
axes = sns.regplot(x=nyc.Date, y=nyc.Temperature)
```

```python
axes = sns.regplot(x=nyc.Date, y=nyc.Temperature)
axes.set_ylim(10, 70)
```

```python
### plot andamento delle temperature plottando anche la linea media negli anni

import numpy as np

fig, ax = plt.subplots()        # a figure with a single Axes
ax.plot(nyc.Date, nyc.Temperature); # plot temp vs date

temp_mean = pd.Series(nyc.Temperature.mean(), range(nyc.Temperature.size)) ## creo una serie di el pari alla media

ax.plot(nyc.Date, temp_mean);
```
