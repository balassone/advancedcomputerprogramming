```python
from matplotlib import pyplot as plt
import numpy as np
fig = plt.figure()
ax = plt.axes()
ax.grid()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
ax.plot(x,(np.cos(x)))
ax.set_title("Multiple");
ax.set_xlabel("x")
ax.set_ylabel("sin e cos allo tan")
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x), label='sin')
ax.plot(x,(np.cos(x)), label='cos')
ax.set_title("Multiple");
ax.set_xlabel("x")
ax.set_ylabel("sin e cos allo tan")
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',color='red')
ax.plot(x,np.cos(x),label='cos',color='green')
ax.set_title('Multiple Lines');
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()
ax.legend()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',color='red',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',color='g',linestyle='dotted')
ax.plot(x,np.tan(x),label='tan',color='b',linestyle='dashdot')
ax.set_title('Multiple Lines');
ax.set_xlabel('x label')          
ax.set_ylim(-5,5)
ax.set_ylabel('y label')
ax.grid()
ax.legend();
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,6))
x = np.linspace(0,10,1000)

axes[0,0].plot(x,np.sin(x))
axes[0,1].plot(x,np.cos(x), color='maroon')
axes[1,0].plot(x, np.sin(x+2), color = 'blue')

axes[1,1].plot(x, np.sin(x+4), color = 'green');
```

# Bar plots

```python
data = {'Apple': 10, 'Orange': 15, 'Lemon': 5, 'Lime': 20}
names = list(data.keys())
values = list(data.values())

fig = plt.figure()
ax = plt.axes()
ax.bar(names,values);
```

```python
data = {'Apple': 10, 'Orange': 15, 'Lemon': 5, 'Lime': 20}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(4,3))
ax = plt.axes()
ax.grid()
ax.barh(names,values,color='teal');
ax.set_xlabel('Fruit')

ax.set_ylabel('Quantity')
ax.set_title('Fruit Quantities');
```

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
width=0.25
x = np.arange(len(labels))

fig, ax = plt.subplots()

rects1 = ax.bar(x-width/2, men_means, width, label='Men')
rects2 = ax.bar(x+width/2, women_means, width, label='Women')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend();
```

```python
fig, ax = plt.subplots()
ax.grid(linestyle='--',color='0.75',axis='y')
ax.set_axisbelow(True)

p1=ax.bar(labels, men_means, width, label='Men')
p2 = ax.bar(labels,women_means, width, bottom=men_means, label='Women')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()
ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2)
ax.set_ylim(0,70);
```

```python
### SCATTER PLOTS

x=np.random.randn(20)
y=np.random.randn(20)

fig, ax = plt.subplots()
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y, marker='x');
```

```python
fig, axs = plt.subplots(2,3,sharex=True,sharey=True,figsize=(16,12));

# marker symbol
axs[0, 0].scatter(x, y, s=80, marker=">")
axs[0, 0].set_title("marker='>'")
# marker from TeX
axs[0, 1].scatter(x, y, s=80, marker=r'$\alpha$')
axs[0, 1].set_title(r"marker=r'\$\alpha\$'")

# marker from path
verts = [[-1, -1], [1, -1], [1, 1], [-1, -1]]
axs[0, 2].scatter(x, y, s=80, marker=verts)
axs[0, 2].set_title("marker=verts")

# regular polygon marker
axs[1, 0].scatter(x, y, s=80, marker=(5, 0))
axs[1, 0].set_title("marker=(5, 0)")

# regular star marker
axs[1, 1].scatter(x, y, s=80, marker=(5, 1))
axs[1, 1].set_title("marker=(5, 1)")

# regular asterisk marker
axs[1, 2].scatter(x, y, s=80, marker=(5, 2))
axs[1, 2].set_title("marker=(5, 2)");
```

```python
### uso di palette di color
### colormap ref: https://matplotlib.org/stable/gallery/color/colormap_reference.html

plt.style.use('seaborn-darkgrid')

z1 = np.sqrt(x**2 + y**2)

fig, ax = plt.subplots()
pos = ax.scatter(x, y, c=z1, cmap='cool', marker='3')

fig.colorbar(pos);
```

```python
x = np.random.randn(100)
y = np.random.randn(100)

z1 = np.sqrt(x**2 + y**2)
z2 = np.random.randint(10, 200, size=len(x))

fig, ax = plt.subplots()
pos = ax.scatter(x, y, c=z1, s=z2, alpha = 0.55, cmap='viridis')
fig.colorbar(pos);
```
